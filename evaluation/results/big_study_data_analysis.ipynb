```python
import pandas as pd
import numpy as np
import scipy as sp
import chart_studio.plotly as py
import plotly.express as px
import plotly.graph_objects as go
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def pdf_fig(figure, filename):
    figure = figure.get_figure()
    figure.savefig("../../thesis/img/big-study/" + filename + ".pdf", bbox_inches = "tight")
```

```python
# seaborn styling
sns.set_style("darkgrid")
sns.set_palette(sns.husl_palette(h=.5, s=1))
sns.set_context("talk")

def label_bars_ints(ax):
    for p in ax.patches:
        ax.annotate("%.0i" % np.nan_to_num(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center', fontsize=14, color='gray', xytext=(0, 10),
            textcoords='offset points')

def label_bars_floats(ax):
    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
            ha='center', va='center', fontsize=14, color='gray', xytext=(0, 10),
            textcoords='offset points')

def process_fig(ax, name, technique, int_annotation, hue_title):
    if (int_annotation):
        label_bars_ints(ax)
    else:
        label_bars_floats(ax)
    if (hue_title != ''):
        ax.legend(loc='center left', bbox_to_anchor=(1.02, 0.5), title=hue_title)
    pdf_fig(ax, name + '-' + technique)
    plt.show()
    ax.clear()
```

```python
techniques = ['pbe', 'ir', 'kw', 'rn']
bool_order = [True, False]
```

```python
pbe = pd.read_csv('pbe.csv')
pbe['Technique'] = 'pbe'
ir = pd.read_csv('ir.csv')
ir['Technique'] = 'ir'
kw = pd.read_csv('keyword.csv')
kw['Technique'] = 'kw'
rn = pd.read_csv('random.csv')
rn['Technique'] = 'rn'
frames = [pbe, ir, kw, rn]

result = pd.concat(frames)
result.rename(columns={'Unnamed: 0':'v'}, inplace=True)

def calc_overlapping_lines(row):
    lines_desired = str(row['DesiredTestOutput']).splitlines()
    lines_output = str(row['TestOutput']).splitlines()
    overlap = list((Counter(lines_output) & Counter(lines_desired)).elements())
    row['TruePosCount'] = len(overlap)
    row['FalsePosCount'] = len(lines_output) - len(overlap)
    row['FalseNegCount'] = len(lines_desired) - len(overlap)
    row['OutputCount'] = len(lines_output)
    row['DesiredCount'] = len(lines_desired)
    row['AllDesiredExtracted'] = row['FalseNegCount'] == 0
    return row

result = result.apply(calc_overlapping_lines, axis=1)

result.head()
```

```python
def rtf(result,technique):
    return result[result['Technique'] == technique]
```

```python
# all techniques - extracted desired log part
ax = sns.countplot(x="AllDesiredExtracted", hue="Technique", data=result, order=bool_order)
process_fig(ax, 'success', 'all', True, 'Technique')
```

```python
# separate technique success rate against number of examples
#ax = sns.catplot(x='ExampleCount', col='Technique', hue = 'AllDesiredExtracted',
#                 data=result, kind='count')
def plot_examples_success_technique(technique):
    ax = sns.countplot(x='ExampleCount', hue='AllDesiredExtracted',
                  data=rtf(result,technique), hue_order=bool_order)
    process_fig(ax, 'success-examples', technique, True, 'AllDesiredExtracted')

for technique in techniques:
    plot_examples_success_technique(technique)
```

```python
# pbe: if the extraction failed, why did it fail? (when was no program learned/regex did not match?)
failed = result[(result.AllDesiredExtracted == False) & (result.Technique == 'pbe')]
# failed_pbe = failed
failed.loc[failed.TestOutput.isin(['no program found', 'no extraction found for this input']), 'to'] = failed.TestOutput
failed.loc[~failed.TestOutput.isin(['no program found', 'no extraction found for this input']), 'to'] = 'extraction found'
failed.loc[failed.TestOutput == 'no extraction found for this input'] = 'no extraction found\nfor this input'

ax = sns.countplot(x="to", data=failed)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set_xlabel('')
process_fig(ax, 'failure-reason', 'pbe', True, '')
```

```python
# precision of extraction: how much of the extracted text is actually necessary
result['Precision'] = result['TruePosCount'] / result['OutputCount']

ax = sns.barplot(x='Technique', y='Precision', hue='AllDesiredExtracted', data=result, hue_order=bool_order)
process_fig(ax, 'precision', 'all', False, 'AllDesiredExtracted')
```

```python
# precision one technique against example count
def precision_examples(technique):
    ax = sns.barplot(x='ExampleCount', y='Precision', hue='AllDesiredExtracted', data=rtf(result, technique), hue_order=bool_order)
    process_fig(ax, 'precision', technique, False, 'AllDesiredExtracted')

for technique in techniques:
    precision_examples(technique)
```

```python
# category distribution
pbe_result = rtf(result, 'pbe')
plt.figure(figsize=(16,4))
order = list(set(pbe_result['Categories']))
order.sort()
order.sort(key = len)
ax = sns.countplot(x='Categories', data=pbe_result, order=order)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")
process_fig(ax, 'categories', 'dataset', True, '')
```

```python
def category_count(categories):
    return len(set(categories.split('-')))
```

```python
# single-category-precsion
result['CategoryCount'] = result.apply(lambda x : category_count(x['Categories']), axis=1)
result['SingleCategory'] = result['CategoryCount'] == 1
result_single_category = result[result['SingleCategory']]
ax = sns.barplot(x='Technique', y='Precision', hue='AllDesiredExtracted', data=result_single_category, hue_order=bool_order)
process_fig(ax, 'single-cateogry-precision', 'all', False, 'AllDesiredExtracted')
```

```python
# single category vs. multiple
ax = sns.barplot(x='Technique', y='Precision', hue='SingleCategory', data=result, hue_order=bool_order)
process_fig(ax, 'precision-category-singularity', 'all', False, 'SingleCategory')
```
